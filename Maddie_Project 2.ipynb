# Part 1
Clean and Organize the Data

```python
# First I want to download the data in CSV format into a Pandas dataframe. 
import pandas as pd 

df = pd.read_csv('Data_Project1.csv', header=None)
print(df)
```

```python
#Fix the column headings 
df.columns = df.iloc[1]
df
```

```python
#Select the relevant columns 
df1 = df[['YEAR', 'STATE', 'TYPE OF PRODUCER', 'ENERGY SOURCE              (UNITS)','CONSUMPTION for ELECTRICITY']] 
df1
```

```python
#Need to eliminate rows 0 and 1 and drop nan values
dfx = df1.drop([0,1]) 
df2 = dfx.dropna()
df2
```

```python
#want basic information on dataset
print(df2.info())
```

```python
#Examine State column 
df2['STATE'].unique()
#found that within the "STATE" column there are two different variables for the total of the US due to capitalization error
```

```python
#want to make the 2 us total variable one consistent variable of "US-TOTAL"
df3 = df2.replace('US-Total', 'US-TOTAL') 
df3
```

```python
#As you can see we now only have 1 value for total US
df3['STATE'].unique()
```

```python
# Examine the Type of Producer column 
df3['TYPE OF PRODUCER'].unique()
```

```python
#Examine Energy Source column
df3['ENERGY SOURCE              (UNITS)'].unique()
#'Other Gases' is not consistent 
```

```python
#Want to replace "Billion BTU" with "Billion Btu" to make the variable consistent 
df4 = df3.replace('Other Gases (Billion BTU)', 'Other Gases (Billion Btu)')
```

```python
df4['ENERGY SOURCE              (UNITS)'].unique()
#now the other gases variable is consistent
```

```python
#Examine Year column 
df4['YEAR'].unique()
```

```python
#want to make the 'year' and 'consumption of electricity' variables numeric, right now they are object
df4.dtypes
```

```python
#importing needed packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#changing year variables to numeric
df4['YEAR'] = pd.to_numeric(df4['YEAR'])
```

```python
df4['CONSUMPTION for ELECTRICITY'] = df4['CONSUMPTION for ELECTRICITY'].str.replace(',', '')
df4['CONSUMPTION for ELECTRICITY'] = df4['CONSUMPTION for ELECTRICITY'].str.replace('.', '')
```

```python
df4['CONSUMPTION for ELECTRICITY'] = pd.to_numeric(df4['CONSUMPTION for ELECTRICITY'])
```

```python
df4.dtypes
```

```python
# We would like to see a graph of the Total US Energy Consumption per year, and per energy source to 
#show how the US consumption of different sources has changed over the time span 1990-2018.

dfUSTOTAL = df4.loc[df4['STATE']== 'US-TOTAL']
#Create a graph of the whole US consumption of energy sources overtime. 
df001 = dfUSTOTAL.loc[df['TYPE OF PRODUCER']=='Total Electric Power Industry']
df000 = df001.pivot(index='YEAR', columns='ENERGY SOURCE              (UNITS)', values='CONSUMPTION for ELECTRICITY')
df000.plot()
```
