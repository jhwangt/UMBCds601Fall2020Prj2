```python
# importing the libraries needed for the EDA of the dataset

import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing the csv file into dataframe

data = pd.read_csv("https://salaries.news.baltimoresun.com/static/csv/cy2017-md.csv")
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
# finding how many duplicate rows in the data

duplicate_rows_data = data[data.duplicated()]
print('number of duplicate rows: ', duplicate_rows_data.shape)
```

```python
# deleting duplicate rows

data = data.drop_duplicates()
```

```python
data.count()
```

```python
# checking how many columns has null values

print (data.isnull().sum())
```

Only Middle Initial and suffix columns have the null values. Which seems okay because not everyone have that.

So, we will fill 0 in the dataset.

```python
#fill NAN values with 0

data = data.fillna(0)
```

```python
#checking the sum of NAN vlaues after filling with 0

data.isnull().sum()
```

```python
# We will plot a bar chart to know which organization has highest number of employees.

data.organization.value_counts().plot(kind='bar', figsize=(18,9))
plt.title("Number of employees by Organization")
plt.ylabel('Number of Employees')
plt.xlabel('Organization')
data.organization.value_counts()
```

```python
# scatterplot of annual salary 
plt.figure(figsize=(9,8))
plt.title("Scatterplot of Annual salary")
plt.ylabel('Amount')
plt.xlabel('Employees')
data['annual_salary'].plot()
```

# Employee with the highest annual salary

```python
#checking who has the maximum salary
max_salary = data[data.annual_salary == data.annual_salary.max()] 
max_salary
```

```python
import seaborn as sns

%matplotlib inline

plt.style.use('bmh')
```

```python
#See the statistics
print(data['annual_salary'].describe()) 

#Plot annual salary distribution
plt.figure(figsize=(9, 8))
plt.xlim(0,500000)
sns.distplot(data['annual_salary'], color='b', bins=100, hist_kws={'alpha': 0.4});
```

```python
#Gross salary is statistics
print(data['ytd_gross_earnings'].describe()) 

#plot ytd grosss salary
plt.figure(figsize=(9, 8))
plt.xlim(0,600000)
sns.distplot(data['ytd_gross_earnings'], color='b', bins=100, hist_kws={'alpha': .3}); 
```

```python
plt.figure(figsize=(12,8))
c= data.corr()
sns.heatmap(c,cmap="RdBu_r",annot=True)
c
```

This graph shows the correlation between the entities which have integer values.

We don't need agency and class_code in the correlation data.

So, let's remove that and create a new list only containing the columns we need.

```python
numbers = list(['annual_salary','regular_earnings','overtime_earnings','other_earnings','ytd_gross_earnings'])
plt.figure(figsize=(12,8))
z = data[numbers].corr()
sns.heatmap(z,cmap="BuPu",annot=True)
z
```
