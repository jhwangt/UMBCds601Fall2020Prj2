```python
#import the required modules
import pandas as pd
import datetime
import datetime as dt
import matplotlib.pylab as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
#read the csv file content with pandas and assign it to data variable
#https://raw.githubusercontent.com/rvkkrishna/DataScience/main/udemy_output_All_Finance__Accounting_p1_p626.csv
data = pd.read_csv('https://raw.githubusercontent.com/rvkkrishna/DataScience/main/udemy_output_All_Finance__Accounting_p1_p626.csv')
```

```python
#Initial Columns
data.columns
```

```python
#Drop the unnecessary Columns
data.drop(['url','discount_price__price_string','price_detail__currency','price_detail__price_string','discount_price__currency'],inplace=True,axis=1)
```

```python
data.columns
```

```python
#heatmap that visualizes the null(NaN) values in the data frame
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

```python
data.fillna(0, inplace=True) #filling the NaN values with 0 as all the NaN are price related
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

```python
data['created'].head() #Created is of object data type, so converting it to datetimestamp
```

```python
#Now created Data is a datatimestamp datatype
data["created Date"] = pd.to_datetime(data["created"],format="%Y-%m-%dT%H:%M:%SZ")
```

```python
data["created Date"].head()
```

```python
#Similarly the published_time(object data type) is converted to Published Date datetimestamp
data["Published Date"] = pd.to_datetime(data["published_time"],format="%Y-%m-%dT%H:%M:%SZ")
```

```python
data['Published Date'].head() #Published is an object, so have to convert to datetimestamp
```

```python
data.drop(['created','published_time'],inplace=True,axis=1) #dropping the created & Published_time as these are objects
```

```python
data.info()  #We observe all the fields datatypes as int or float or datetime or bool
```

```python
# To fetch and plot the subscribers over the years from 2010 to 2020
years = [year for year, data in data.groupby(['Published Year'])]
plt.bar(years,data.groupby(['Published Year']).sum()['num_subscribers'])
plt.ylabel('Subscribers')
plt.xlabel('Years')
plt.xticks(years, rotation='vertical', size=10)
plt.show()
```

```python
#Ploting Paid and Free courses over total Udemy courses offered 

mypie = data.is_paid.value_counts()
mypie.rename({True: 'Paid', False: 'Free'}, inplace=True)

fig = go.Figure(data=[go.Pie(labels=mypie.index, values=mypie, hole=.4, textinfo='label+percent')])
fig.update_layout(title_text="Free/paid courses")
fig.show()
```

```python
per = data['Published Date'].dt.to_period("M")
g = data.groupby(per)
coldata = g.sum()
```

```python
fig = plt.figure(figsize=(20,10))
coldata['num_subscribers'].plot(kind="bar",figsize=(100,6))

plt.setp(plt.gca().get_xticklabels(), rotation=60, fontsize=10)
plt.title("The No. Of Subscribers Over the Months For Each Year")
plt.xlabel('Date', fontsize=20)
plt.ylabel('No. Of Subscribers', fontsize=20)
```

```python
#Courses with maximum no of subscribers

dd = data.nlargest(10,'num_subscribers')

fig = plt.figure(figsize = (20, 5)) 
  
# creating the bar plot 
plt.bar(dd['title'], dd['num_subscribers'], color ='maroon', width = 0.4) 

plt.xlabel("Courses Names") 
plt.ylabel("No. Of Subscribers") 
plt.title("Courses With Top 10 Subscribers") 
plt.xticks(rotation=90)

plt.show()
```

```python
#Courses With Top 50 Max no. of Published Lecturers

dd = data.nlargest(50,'num_published_lectures')

fig = plt.figure(figsize = (20, 10)) 
  
# creating the bar plot 
plt.bar(dd['title'], dd['num_published_lectures'], color ='maroon', width = 0.4) 

plt.xlabel("Courses Names") 
plt.ylabel("No. Of Published Lecturers") 
plt.title("Courses With Top 50 Max no. of Published Lecturers") 
plt.xticks(rotation=90)

plt.show()
```

```python
# To fetch and plot the top 30 courses offering max discount
data['Max_Discount_Offerred']=data['discount_price__amount']/data['price_detail__amount'] *100
dd = data.nlargest(30,'Max_Discount_Offerred')


fig = plt.figure(figsize = (20, 10)) 
  
# creating the bar plot 
plt.bar(dd['title'], dd['Max_Discount_Offerred'], color ='maroon', width = 0.4) 

plt.xlabel("Courses Names") 
plt.ylabel("Discount Percentage Offerred ") 
plt.title("Courses With Top 30 Max Discount Courses") 
plt.xticks(rotation=90)

plt.show()
```
